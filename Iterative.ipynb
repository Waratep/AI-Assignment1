```python
def checkwall(map1,current):
    x=current[0]
    y=current[1]
    if(map1[x][y] != 9):
        return True
    return False
```

```python
def dfs_paths(graph, start, goal):
    
    stack = [(start, [start])]
    while stack:
        (vertex, path) = stack.pop()
        for next in graph[vertex] - set(path):
            if next == goal:
                yield path + [next]
                
            else:
                stack.append((next, path + [next]))
```
